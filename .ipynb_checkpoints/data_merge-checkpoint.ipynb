# Merge


1. seen_and_click = seen **union** click
2. is_read = **add** column is_read to read
3. seen_click_read = seen_and_click **leftjoin** is_read
4. merge seen_click_read with action
5. merge with project_table

## 1. seen_and_click = seen **union** click


```python
seen.shape
```

```python
click.shape
```

```python
seen_and_click = pd.merge(seen, click,  how='outer', left_on=['article_id','user_type','user_id','project_id'], right_on = ['article_id','user_type','user_id','project_id'])
```

```python
print(seen_and_click.shape)
seen_and_click[0:20]
```

```python
len(seen_and_click.groupby(['article_id','user_type','user_id','project_id']).groups)
```

```python
new_seen_and_click =  seen_and_click[~(seen_and_click['seen_at'] > seen_and_click['clicked_at']) ]
print(new_seen_and_click.shape)
new_seen_and_click.head()
```

```python
new_seen_and_click = new_seen_and_click.drop_duplicates(subset=['article_id', 'project_id', 'user_id', 'user_type'], keep='first')
print(new_seen_and_click.shape)
new_seen_and_click.head() 
```

## 2. is_read = **add** column is_read to read

```python
print(read.shape)
read.head()
```

```python
print(article.shape)
article.head()
```

```python
new_read = pd.merge(read, article[['id','read_time']], how='left', left_on=['article_id'], right_on = ['id'])
print(new_read.shape)
new_read[0:10] 
```

```python
new_read['is_read'] = ((new_read['read_time_x']/new_read['read_time_y'] > 0.7) & (new_read['read_length'] > 70)).astype(int)
```

```python
if 'id' in new_read:
    del new_read['id'] 
if 'read_time_y' in new_read:
    del new_read['read_time_y']
new_read = new_read.rename(columns={'read_time_x':'read_time'})
print(new_read.shape)
new_read.head() 
```

```python
new_read = new_read.drop_duplicates(subset=['article_id', 'project_id', 'user_id', 'user_type'], keep='first')
print(new_read.shape)
new_read.head() 
```

# 3. seen_click_read = seen_and_click **leftjoin** is_read

```python
print(new_seen_and_click.shape)
new_seen_and_click.head()
```

```python
seen_click_read = pd.merge(new_seen_and_click, new_read, how='left', left_on=['article_id','project_id','user_id','user_type'], right_on = ['article_id','project_id','user_id','user_type'])
seen_click_read[0:10]
```

# 4. merge seen_click_read with action

```python
print(action.shape)
action.head()
```

```python
new_action = action.drop_duplicates(subset=['article_id', 'project_id', 'user_id', 'user_type'], keep='first')
print(new_action.shape)
new_action.head() 
```

```python
seen_click_read_action = pd.merge(seen_click_read, new_action, how='left', left_on=['article_id','project_id','user_id','user_type'], right_on = ['article_id','project_id','user_id','user_type'])
print(seen_click_read_action.shape)
seen_click_read_action.head()
```

# 5. merge with project_table

```python
seen_click_read_action_project = pd.merge(seen_click_read_action, project, how='left', left_on=['project_id'], right_on = ['id'])

if 'id' in seen_click_read_action_project:
    del seen_click_read_action_project['id'] 
seen_click_read_action_project = seen_click_read_action_project.rename(columns={'name':'Project'})

print(seen_click_read_action_project.shape)
seen_click_read_action_project.head()
```

```python
seen_click_read_action_project.to_csv('data_report/report.csv', index=False)
```
